```python
def fuzzy_age(val):
    result = {"Muda": 0, "Dewasa": 0, "Tua": 0}
    a, b, c, d = 30, 45, 60, 75

    if val <= a:
        result["Muda"] = 1
    elif a < val < b:
        result["Muda"] = (b - val) / (b - a)

    if a < val < b:
        result["Dewasa"] = (val - a) / (b - a)
    elif b <= val <= c:
        result["Dewasa"] = 1
    elif c < val < d:
        result["Dewasa"] = (d - val) / (d - c)

    if val >= d:
        result["Tua"] = 1
    elif c < val < d:
        result["Tua"] = (val - c) / (d - c)

    return result
```

```python
def fuzzy_heartrate(val):
    result = {"Rendah": 0, "Normal": 0, "Tinggi": 0}
    a, b, c, d = 50, 65, 85, 100

    if val <= a:
        result["Rendah"] = 1
    elif a < val < b:
        result["Rendah"] = (b - val) / (b - a)

    if a < val < b:
        result["Normal"] = (val - a) / (b - a)
    elif b <= val <= c:
        result["Normal"] = 1
    elif c < val < d:
        result["Normal"] = (d - val) / (d - c)

    if val >= d:
        result["Tinggi"] = 1
    elif c < val < d:
        result["Tinggi"] = (val - c) / (d - c)

    return result
```

```python
def fuzzy_ckmb(val):
    result = {"Rendah": 0, "Sedang": 0, "Tinggi": 0}
    a, b, c, d = 1.5, 5, 10, 20

    if val <= a:
        result["Rendah"] = 1
    elif a < val < b:
        result["Rendah"] = (b - val) / (b - a)

    if a < val < b:
        result["Sedang"] = (val - a) / (b - a)
    elif b <= val <= c:
        result["Sedang"] = 1
    elif c < val < d:
        result["Sedang"] = (d - val) / (d - c)

    if val >= d:
        result["Tinggi"] = 1
    elif c < val < d:
        result["Tinggi"] = (val - c) / (d - c)

    return result
```

```python
def fuzzy_troponin(val):
    result = {"Rendah": 0, "Sedang": 0, "Tinggi": 0}
    a, b, c, d = 0.1, 0.3, 0.6, 1.0

    # Rendah
    if val <= a:
        result['Rendah'] = 1
    elif a < val < b:
        result['Rendah'] = (b - val) / (b - a)

    # Sedang
    if a < val < b:
        result['Sedang'] = (val - a) / (b - a)
    elif b <= val <= c:
        result['Sedang'] = 1
    elif c < val < d:
        result['Sedang'] = (d - val) / (d - c)

    # Tinggi
    if val >= d:
        result['Tinggi'] = 1
    elif c < val < d:
        result['Tinggi'] = (val - c) / (d - c)

    return result
```
